```python
#EJEMPLO DE LIMPIEZA Y TRANSFORMACIÓN

# Aquí voy a desnormalizar las tablas gameAppearance, Characters y games

# Al llamar la tabla completa algunas series se repiten: character id con id_x, game_title con title, role_x con role_y, game_id con id_y por lo que se implementará un método Drop para limpiar las series del DataFrame

import pandas as pd

ruta_games = '../data/raw/games.csv'
ruta_gameAppear = '../data/raw/gameAppearance.csv'
ruta_char = '../data/raw/characters.csv'

df_gameAppear = pd.read_csv(ruta_gameAppear)
df_games = pd.read_csv(ruta_games)
df_char = pd.read_csv(ruta_char)


df_union1 = pd.merge(
    df_gameAppear,
    df_char,
    left_on='character_id',
    right_on='id',
    how='left'
)

df_union_completa = pd.merge(
    df_union1,
    df_games,
    left_on='game_id',
    right_on='id',
    how='left'
)

df_union_completa #Nos muestra la tabla inicial sin cambios

columnas_a_borrar = ['id_x','id_y','role_y','title','name'] # Define la lista de las series a eliminar
df_limpio = df_union_completa.drop(columns=columnas_a_borrar) #Define la función con el método drop

# Así podemos remplazar los nombres de una serie
df_limpio = df_limpio.rename(columns={
    'name_x': 'character_name',
    'role_x': 'role_in_game'
})

df_limpio #Ahora tenemos las tres tablas desnormalizadas; en esta tabla podemos ver todos los personajes (Principales y no princiaples) y las apariciones que tienen en los juegos originales y su cronología
```
