```python
# Visualización de los datos para la tabla df_limpio

#Personajes con más apariciones

import matplotlib.pyplot as plt
import pandas as pd

ruta_procesada = '..\data\processed\df_limpio.csv'
df = pd.read_csv(ruta_procesada)
df # Checado de la tabla

top_personajes = df['character_name'].value_counts().head(10) #Variable para sacar el numero de apariciones en un rango de 10
top_personajes = top_personajes[::-1] #Esto hace que nuestra tabla se invierta y pueda salir el personaje de más apariciones al que menos tiene

plt.barh(y=top_personajes.index, width=top_personajes.values) #Método para realizar una grafica de barras acostada
plt.ylabel('Nombre del personaje') 
plt.xlabel('Apariciones')
plt.title('Top personajes con más apariciones')

plt.show()





```

```python
# Visualización de los datos para la tabla df_limpio

#Cronología de los juegos de lanzamiento de los juegos

import matplotlib.pyplot as plt
import pandas as pd

ruta_procesada = '..\data\processed\df_limpio.csv'
df = pd.read_csv(ruta_procesada)
df # Checado de la tabla

cronologia = df[['game_title', 'year']].drop_duplicates().sort_values('year')

plt.scatter(x=cronologia['year'], y=cronologia['game_title'])

plt.title('Cronología de Resident Evil')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Titulo del juego')

plt.show()
```
